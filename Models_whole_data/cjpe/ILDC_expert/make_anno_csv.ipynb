```python
# Specifica la cartella di origine dei file di testo
cartella_origine = "/content/drive/MyDrive/Thesis/Models_whole_data/case_explanation/ILDC_expert/source"

# Specifica la cartella di destinazione per i file modificati
cartella_destinazione = "/content/drive/MyDrive/Thesis/Models_whole_data/case_explanation/ILDC_expert/source_new"

# Verifica che la cartella di origine esista
if os.path.exists(cartella_origine):
    # Crea la cartella di destinazione se non esiste
    if not os.path.exists(cartella_destinazione):
        os.makedirs(cartella_destinazione)

    # Elabora ciascun file nella cartella di origine
    for nome_file in os.listdir(cartella_origine):
        percorso_origine = os.path.join(cartella_origine, nome_file)
        percorso_destinazione = os.path.join(cartella_destinazione, nome_file)

        # Verifica se il percorso è un file
        if os.path.isfile(percorso_origine):
            with open(percorso_origine, 'r') as file:
                linee = file.readlines()

            # Verifica se ci sono almeno due righe nel file
            if len(linee) >= 2:
                # Rimuovi l'ultima riga
                linee = linee[:-1]

                # Scrivi il file modificato nella cartella di destinazione
                with open(percorso_destinazione, 'w') as file:
                    file.writelines(linee)
                print(f"Ultima riga rimossa da {nome_file} e salvata in {percorso_destinazione}")
            else:
                print(f"Il file {nome_file} ha meno di due righe, non è stato modificato.")
        else:
            print(f"{nome_file} non è un file.")
else:
    print(f"La cartella di origine {cartella_origine} non esiste.")
```

```python
import os
import pandas as pd

# Definisci la cartella in cui si trovano i file TXT
cartella = "/content/drive/MyDrive/Thesis/Models_whole_data/case_explanation/ILDC_expert/source_new"

# Crea due liste vuote per i dati e i nomi dei file
dati = []
nomi_file = []

# Scorri tutti i file nella cartella
for filename in os.listdir(cartella):
    if filename.endswith(".txt"):
        file_path = os.path.join(cartella, filename)
        with open(file_path, "r") as file:
            # Leggi il contenuto del file
            file_content = file.read()
            dati.append(file_content)
            nomi_file.append(filename)

# Crea un DataFrame pandas con due colonne: "NomeFile" e "Contenuto"
df = pd.DataFrame({"text": dati, "name": nomi_file})

# Ora il DataFrame df contiene il nome del file in una colonna separata.
```
